# 3) EXPLORATORY DATA ANALYSIS (EDA) with Visualization and appropriate insights:

# 3.1) Accident Severity Distribution

# 3.9) WEEKDAY vs WEEKEND ACCIDENTS

```python
weekend_days = ['Saturday', 'Sunday']

rta['Day_Type'] = rta['Day_of_week'].apply(
    lambda x: 'Weekend' if x in weekend_days else 'Weekday'
)
```

```python
daytype_counts = rta['Day_Type'].value_counts()      # To show accident counts on weekdays and on wekends
print(daytype_counts)
```

```python
plt.figure()
daytype_counts.plot(kind='bar')
plt.title("Weekday vs Weekend Accidents")    # Bar Chart which shows weekdays vs weekend accidents for Hour Day vs the Accident Counts.
plt.xlabel("Day Type")
plt.ylabel("Number of Accidents")
plt.show()
```

```python
rta['Time'] = pd.to_datetime(rta['Time'], errors='coerce')
rta['Hour'] = rta['Time'].dt.hour

weekday_map = {
    'Monday':'Weekday', 'Tuesday':'Weekday', 'Wednesday':'Weekday',  # Line Chart Representation for detailed insights
    'Thursday':'Weekday', 'Friday':'Weekday',
    'Saturday':'Weekend', 'Sunday':'Weekend'
}
rta['Day_type'] = rta['Day_of_week'].map(weekday_map)
hour_daytype = rta.groupby(['Hour','Day_type']).size().reset_index(name='Accident_Count')

pivot = hour_daytype.pivot(index='Hour', columns='Day_type', values='Accident_Count')

pivot.plot(kind='line', marker='o', figsize=(10,6))
plt.title("Hour-wise Accident Distribution: Weekday vs Weekend")
plt.xlabel("Hour of Day")
plt.ylabel("Accident Count")
plt.xticks(range(0,24,4))
plt.legend(title="Day Type")
plt.grid(True, linestyle="--", alpha=0.6)
plt.show()
```

# 4.2) Casualties Number vs the Weather Condition

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(1, 2, figsize=(12,5))

# Boxplot: Casualties by Weather Conditions
rta.boxplot(column='Number_of_casualties', by='Weather_conditions', ax=axes[0])
axes[0].set_title("Boxplot: Casualties by Weather")
axes[0].set_xlabel("Weather Conditions")
axes[0].set_ylabel("Number of Casualties")
axes[0].tick_params(axis='x', rotation=30)

# Bar plot: Total Casualties by Weather Conditions
weather_totals = rta.groupby('Weather_conditions')['Number_of_casualties'].sum()
axes[1].bar(weather_totals.index, weather_totals.values)
axes[1].set_title("Bar Plot: Total Casualties by Weather")
axes[1].set_xlabel("Weather Conditions")
axes[1].set_ylabel("Total Casualties")
axes[1].tick_params(axis='x', rotation=30)

plt.suptitle("Casualties vs Weather Conditions", fontsize=14)
plt.tight_layout()
plt.show()
```

# Insights:
- Normal Weather Has the Highest Total Casualties
From the bar chart:
Normal weather conditions contribute the highest total casualties around 15000 or more
Followed by Raining around 2200 or more.
Other conditions (Cloudy, Snow, Windy, Fog) contribute comparatively fewer casualties
This does NOT mean normal weather is more dangerous.
It likely means:
Most driving happens under normal weather.
Higher exposure → higher total accidents → higher total casualties.
- Rainy Conditions Show Higher Risk Per Accident
From the boxplot:
Rain shows slightly higher variability
Some outliers with 4–6 casualties per accident
Mean casualties appear slightly higher compared to normal weather
This suggests rainy weather may increase accident severity per incident, even though total cases are lower than normal weather.
- Most Accidents Result in 1–2 Casualties
Across almost all weather conditions:
Median ≈ 1 casualty
Majority of accidents involve 1 or 2 victims
Severe multi-casualty accidents (5+) are rare (shown as outliers)
